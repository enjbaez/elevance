```python
import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as plt
```

```python
# plot the original vector v
v1 = np.array([1,1])
plt.plot([0, v1[0]],[0,v1[1]], 'b',label= 'v', linewidth=2)
plt.plot([0,1],[0,0],'r:', label = 'ihat')
plt.plot([0,0],[0,1],'r:', label ='jhat')

plt.legend()
plt.axis("square")
plt.axis((-0.5,1.5,-0.5,1.5))
plt.grid()
plt.show()
```

```python
#Operations
#scale
#rotate
#sheer
```

```python
# original vector
v1 = np.array([1,1])

#scale (different numbers along the diagonal transform matrix)
# 2 0
# 0 1
TFM_Scale = np.array([[2,0],[0,3]])
v2 = TFM_Scale @ v1
print("Scale:")
print(TFM_Scale)
print(v2)
```

```python
# rotate
TFM_Rotate = np.array([[0,-1],[1,0]])
v3 = TFM_Rotate @ v1
print("Rotate:")
print(TFM_Rotate)
print(v3)
```

```python
# shear (along the x axis)
TFM_Shear = np.array([[1,1],[0,1]])
v4 = TFM_Shear @ v1
print("Shear:")
print(TFM_Shear)
print(v4)

plt.plot([0, v1[0]],[0, v1[1]], 'b', label ='v', linewidth=2)
plt.plot([0, v2[0]],[0, v2[1]], 'r', label = 'scale', linewidth=2)
plt.plot([0, v3[0]],[0, v3[1]], 'g', label = 'rotate', linewidth=2)
plt.plot([0, v4[0]],[0, v4[1]], 'y', label = 'shear', linewidth=2)

plt.legend()
plt.axis("square")
plt.axis((-2,3,-1,4))
plt.grid()
plt.show()
```

```python
# The order of the transformation matters
# scale, rotate, shear
v5 = TFM_Scale @ v1
v6 = TFM_Rotate  @ v5
v7 = TFM_Shear @ v6

# rotate, shear, scale
v8 = TFM_Rotate @ v1
v9 = TFM_Shear @ v8
v10 = TFM_Scale @ v9

#plot the transformation 
plt.plot([0, v1[0]], [0, v1[1]], 'b', label='v', linewidth=2)
plt.plot([0, v7[0]], [0, v7[1]], 'g:', label='sc/rot/sh', linewidth=2)
plt.plot([0, v10[0]], [0, v10[1]], 'r:', label='ro/sh/sc', linewidth=2)

plt.legend(loc = 'upper right')
plt.axis('square')
plt.axis((-2,3,-1,4))
plt.grid
plt.show
```
