```python
import altair as alt
import pandas as pd
```

```python
wide_form = pd.DataFrame({'Year':[2018,2019,2020],
                          'Clothing':[110,101,129],
                          'Furniture':[201,184,195],
                          'Health':[120,132,133]
                         })
wide_form
```

```python
long_form = pd.DataFrame({'Year':[2018, 2019,2020,
                                  2018,2019,2020,
                                  2018,2019,2020],
                          'Categories': ['Clothing', 'Clothing', 'Clothing',
                                         'Furniture', 'Furniture', 'Furniture',
                                         'Health', 'Health', 'Health'],
                          'Sales':[110,101,129,
                                   201,184,195,
                                   120,132,133]
                         })
long_form
```

```python
alt.Chart(
    long_form).mark_line().encode(
    x='Year:O',
    y='Sales:Q',
    color= 'Category:N'
).properties(
    height = 400,
    width = 600,
    title='Annual Sales by Product Category')
```

```python
wide_form.melt('Year',var_name='Category',value_name='Sales')
```

```python
alt.Chart(wide_form.melt('Year',var_name='Category',value_name='Sales')
         ).mark_line().encode(
            x='Year:O',
            y='Sales:O',
            color='Category:N'
).properties(
    height = 400,
    width = 600,
    title='Annual Sales by Presuct Category')
            
```
