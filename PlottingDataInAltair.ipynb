```python
import altair as alt
import pandas as pd
```

```python
data = pd.DataFrame({'Categories': ['Clothing',
                                    'Electronics',
                                    'Furniture',
                                    'Health',
                                    'Kitchen'],
                     'Sales': [100,343,201,120,278]})
data
```

```python
alt.Chart(data).mark_bar().encode(
    x='Categories',
    y='Sales',
)
```

```python
alt.Chart(
    data,
    height = 400,
    width = 600
).mark_bar().encode(
    x='Categories',
    y='Sales',
)
```

```python
alt.Chart(data).mark_bar().encode(
    x='Categories',
    y='Sales'
).properties(
    height = 400,
    width = 600)
```

```python
alt.Chart(data).mark_bar().encode(
    x='Categories',
    y='Sales'
).properties(
    height = 400,
    width = 600,
    title='Sales by Product Category')
```

```python
alt.Chart(data).mark_bar().encode(
    x='Categories:O',
    y='Sales:Q'
).properties(
    height = 400,
    width = 600,
    title='Sales by Product Category')
```

```python
alt.Chart(data).mark_bar().encode(
    x='Categories:Q',
    y='Sales:O'
).properties(
    height = 400,
    width = 600,
    title='Sales by Product Category')
```
