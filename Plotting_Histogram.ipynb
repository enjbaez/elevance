```python
import altair as alt
import pandas as pd
```

```python
cars_data = pd.read_csv("datasets/CarPrice_Assignment.csv")
```

```python
cars_data.head()
```

```python
cars_data.shape
```

```python
cars_data['horsepower'].describe()
```

```python
alt.Chart(cars_data).mark_bar().encode(
    x = "horsepower", y = "count()"
).properties(width=600)
```

```python
alt.Chart(cars_data).mark_bar().encode(
    x=alt.X('horsepower:Q', bin=True),
    y = 'count()'
).properties(width=600)
```

```python
alt.Chart(cars_data).mark_bar().encode(
    x=alt.X('horsepower:Q', bin=alt.BinParams(maxbins=3)),
    y='count()'
).properties(width=600)
```

```python
cars_data['enginesize'].describe()
```

```python
alt.Chart(cars_data).mark_bar().encode(
    x=alt.X('enginesize:Q', bin=True),
    y='count()'
).properties(width=600)
```

```python
alt.Chart(cars_data.transform_fold(
    ['enginesize']
).mark_bar().encode(
    x=alt.X('value:Q', bin=True),
    y='count()'
).properties(width=600)
```

```python
alt.Chart(cars_data).transform_fold(
    ['enginesize'],
    as_=['enginevol', 'Engine Size (cu. inches)'],
).mark_bar().encode(
    x=alt.X('Engine Size (cu. inches):Q', bin=True),
    y = 'count()'
).properties(width=600)
```

```python
alt.Chart(cars_data).transform_fold(
    ['enginesize'],
).transform_density(
    density='value'
).mark_bar().encode(
    alt.X('value:Q',bin=alt.BinParams(maxbins=10)),
    alt.Y('density:Q')
).properties(width=600)
```
